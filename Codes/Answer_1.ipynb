### Importing libraries that will be useful :-


```python
import json
import pandas as pd 
import numpy as np
from pprint import pprint
```

# Question 1 solution : 

```python
# Loading data from neighbor-district.json file
ptr = open('neighbor-districts.json')
neigh_dis = json.load(ptr)
pprint(neigh_dis)
# neigh_dis
```

## Cleaning data :-



```python
# removing all the contents after / from keys like "/Q15732396" 
modified_neigh_dis =  { k.split('/')[0]: v for k, v in neigh_dis.items() } 

# replacing "_" with space from keys
modified_neigh_dis =  { k.replace("_"," "): v for k, v in modified_neigh_dis.items() }

# removing "_district" from keys like ahmedabad_district
modified_neigh_dis =  { k.replace(" district",""): v for k, v in modified_neigh_dis.items() }

# removing all the contents after / from values like "/Q15732396" 
for key, value in modified_neigh_dis.items():
    modified_neigh_dis[key] = list(map(lambda x: x.split('/')[0], value))
    
# replacing "_" with space from values
for key, value in modified_neigh_dis.items():
    modified_neigh_dis[key] = list(map(lambda x: x.replace("_"," "), value))
    
# removing "_district" from values like anand_district
for key, value in modified_neigh_dis.items():
    modified_neigh_dis[key] = list(map(lambda x: x.replace(" district",""), value))
    
# Finally printing our cleaned data 
# pprint(modified_neigh_dis)
# len(modified_neigh_dis) is 717 
```

## Creating a dictionary of list with corrected district names (basically pre-final output file)

```python
# Written by Anuj 
pre_final = {}
for key, value in modified_neigh_dis.items():
    abc = []
    for i in value:
        abc.append(corrected_districts.get(i, i))
    a = corrected_districts.get(key, key)
    pre_final[a] = abc
# len(pre_final) # = 717
# pre_final
```

```python
### Removing districts that are not in any of the covid data and vaccine data. ###

del pre_final['noklak']
del pre_final['niwari']
del pre_final['mumbai suburban']
del pre_final['medchalmalkajgiri']
del pre_final['konkan division']
del pre_final['janjgir-champa']
```

```python
# len(pre_final) # = 711
```

## Finally creating neighbor-districts-modified.json i.e the final output file

```python
final = {}
for key, value in pre_final.items():
    abc = []
    for i in value:
        abc.append(common_districts_dict_keys.get(i, i)) # This we are doing to create a dict of list with districts names
    a = common_districts_dict_keys.get(key, key)         #  replaced by their district ids...
    final[a] = abc
# len(final) # = 711
# final
```

### Writing final dict output into neighbor-districts-modified.json filem

```python
with open("neighbor-districts-modified.json", "w") as temp:
    json.dump(final, temp, sort_keys=True, indent=4)
```
